## Q3 Show minimum and maximum bloodpressure of diabetic patient who smokes. Make column names as MinBP and MaxBP respectively.

```python
# use this query in sql workbench
# SELECT MIN(bloodpressure) AS MinBP, MAX(bloodpressure) AS MaxBP 
# FROM insurance_data 
# WHERE diabetic='yes' AND smoker='yes';
```

```python
# create a cursor object
cursor = con.cursor()

# execute the SQL query
cursor.execute("SELECT MIN(bloodpressure) AS MinBP, MAX(bloodpressure) AS MaxBP FROM insurance_data WHERE diabetic='yes' AND smoker='yes'")

# fetch the result
result = cursor.fetchone()

# print the result
print("Minimum Blood Pressure:", result[0])
print("Maximum Blood Pressure:", result[1])

# # close the cursor and connection
# cursor.close()
# con.close()
```

## Q4 Find no of unique patients who are not from southwest region.

## Q7 No of patient having normal blood pressure. Normal range[90-120]

```python
# use this query in sql workbench
# SELECT COUNT(*) as normal_bp_count 
# FROM insurance_data 
# WHERE bloodpressure >= 90 AND bloodpressure <= 120;
```

## Q8 No of pateint below 17 years of age having normal blood pressure as per below formula -

### BP normal range = 80+(age in years × 2) to 100 + (age in years × 2)

```python
# use this query in sql workbench
# SELECT COUNT(*) as normal_bp_count 
# FROM insurance_data 
# WHERE age < 17 AND bloodpressure >= (80 + (age * 2)) AND bloodpressure <= (100 + (age * 2));
```

```python
# create a cursor object
cursor = con.cursor()

# execute the SQL query
cursor.execute("SELECT COUNT(*) as normal_bp_count FROM insurance_data WHERE age < 17 AND bloodpressure >= (80 + (age * 2)) AND bloodpressure <= (100 + (age * 2))")

# fetch the result
result = cursor.fetchone()

# print the result
print("Number of patients below 17 years of age with normal blood pressure:", result[0])

# # close the cursor and connection
# cursor.close()
# con.close()
```

## Q9 What is the average claim amount for non-smoking female patients who are diabetic?

```python
SELECT AVG(claim) AS avg_claim_amount
FROM insurance_data 
WHERE gender = 'female' AND smoker = 'no' AND diabetic = 'yes';
```

```python
# create a cursor object
cursor = con.cursor()

# execute the SQL query
cursor.execute("SELECT round(AVG(claim),2) AS avg_claim_amount FROM insurance_data WHERE gender = 'female' AND smoker = 'no' AND diabetic = 'yes'")

# fetch the result
result = cursor.fetchone()

# print the result
print("Average claim amount for non-smoking female patients who are diabetic:", result[0])

# # close the cursor and connection
# cursor.close()
# con.close()
```
