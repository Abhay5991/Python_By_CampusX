### Problem 4: Group by on exercise frequency and smoking status and show average deep sleep time, average light sleep time and avg rem sleep time.
### Note the differences in deep sleep time for smoking and non smoking status

```python
#using sql workbentch
SELECT exercise_frequency, smoking_status,
ROUND(AVG(sleep_duration * deep_sleep_percentage / 100), 2) AS avg_deep_sleep_time,
ROUND(AVG(sleep_duration * light_sleep_percentage / 100), 2) AS avg_light_sleep_time,
ROUND(AVG(sleep_duration * rem_sleep_percentage / 100), 2) AS avg_rem_sleep_time
FROM Sleep_Efficiency
WHERE deep_sleep_percentage BETWEEN 25 AND 45
GROUP BY exercise_frequency, smoking_status;
```

### Problem 5: Group By on Awekning and show AVG Caffeine consumption, AVG Deep sleep time and AVG Alcohol consumption only for people who do exercise atleast 3 days a week. Show result in descending order awekenings

```python
#using sqlworkbentch
SELECT Awakenings,
    AVG(Caffeine_consumption) as avg_caffeine_consumption,
    AVG(sleep_duration*Deep_sleep_percentage/100) as avg_deep_sleep_time,
    AVG(Alcohol_consumption) as avg_alcohol_consumption
FROM Sleep_Efficiency
WHERE Exercise_frequency >= 3
GROUP BY Awakenings
ORDER BY Awakenings DESC;
```

### **`Problem 6:`**



**The question is:**

Display those power stations which have average 'Monitored Cap.(MW)' (display the values) between 1000 and 2000 and the number of occurance of the power stations (also display these values) are 
greater than 200. Also sort the result in ascending order.




```python
# using sql wokbentch the query is 
SELECT `Power Station`, AVG(`Monitored Cap.(MW)`) AS avg_monitored_cap, COUNT(*) AS num_occurrences
FROM session_33.powergeneration
GROUP BY `Power Station`
HAVING avg_monitored_cap BETWEEN 1000 AND 2000 AND num_occurrences > 200
ORDER BY avg_monitored_cap ASC
```

```python
import pandas as pd
import pymysql

# establish database connection
con = pymysql.connect(host='localhost', user='root', password='campusx', database='session_33')

# execute query and store results in a DataFrame
df = pd.read_sql_query("SELECT `Power Station`, ROUND(AVG(`Monitored Cap.(MW)`),2) AS avg_monitored_cap, COUNT(*) AS num_occurrences \
                        FROM `powergeneration` \
                        GROUP BY `Power Station` \
                        HAVING avg_monitored_cap BETWEEN 1000 AND 2000 AND num_occurrences > 200 \
                        ORDER BY avg_monitored_cap ASC", con)

# display the DataFrame
print(df)
```

### **`Problem 7:`**

**The question is:**

Display top 10 lowest "value" State names of which the Year either belong to 2013 or 2017 or 2021 and type is 'Public In-State'. Also the number of occurance should be between 6 to 10. Display the average value upto 2 decimal places, state names and the occurance of the states.


```python
#using sql workbentch 
SELECT State, ROUND(AVG(Value),2) AS avg_value, COUNT(*) AS num_occurrences
FROM session_33.College_cost
WHERE Year IN (2013, 2017, 2021) AND Type = 'Public In-State'
GROUP BY State
HAVING num_occurrences BETWEEN 6 AND 10
ORDER BY avg_value ASC
LIMIT 10;
```

```python
import pandas as pd
import pymysql

# establish database connection
con = pymysql.connect(host='localhost', user='root', password='campusx', database='session_33')

# execute query and store results in a DataFrame
df = pd.read_sql_query("SELECT State, ROUND(AVG(Value),2) AS avg_value, COUNT(*) AS num_occurrences \
                        FROM College_cost \
                        WHERE Year IN (2013, 2017, 2021) AND Type = 'Public In-State' \
                        GROUP BY State \
                        HAVING num_occurrences BETWEEN 6 AND 10 \
                        ORDER BY avg_value ASC \
                        LIMIT 10", con)

# round the average value to 2 decimal places
df['avg_value'] = df['avg_value'].round(2)

# display the DataFrame
print(df)
```

Display total and average values of Discount_offered for all the combinations of 'Mode_of_Shipment' (display this feature) and 'Warehouse_block' (display this feature also) for all male ('M') and 'High' Product_importance. Also sort the values in descending order of Mode_of_Shipment and ascending order of Warehouse_block.

```python
engine = create_engine("mysql+pymysql://root:campusx@localhost/session_33")
# {root}:{password}@{url}/{database}
shipping_ecommerce.to_sql('shipping_ecommerce', con = engine)
```

### Solution

```python
#using Sql Workbentch
SELECT Mode_of_Shipment, 
    Warehouse_block, 
    SUM(Discount_offered) AS total_discount, 
    ROUND(AVG(Discount_offered),2) AS avg_discount
FROM session_33.shipping_ecommerce
WHERE Gender = 'M' AND Product_importance = 'High'
GROUP BY Mode_of_Shipment, Warehouse_block
ORDER BY Mode_of_Shipment DESC, Warehouse_block ASC;
```

```python
import pandas as pd
import pymysql

# establish database connection
con = pymysql.connect(host='localhost', user='root', password='campusx', database='session_33')

# execute query and store result in a DataFrame
df = pd.read_sql_query("SELECT Mode_of_Shipment, Warehouse_block, SUM(Discount_offered) AS total_discount, ROUND(AVG(Discount_offered),2) AS avg_discount \
                        FROM shipping_ecommerce \
                        WHERE Gender = 'M' AND Product_importance = 'High' \
                        GROUP BY Mode_of_Shipment, Warehouse_block \
                        ORDER BY Mode_of_Shipment DESC, Warehouse_block ASC", con)

# display the result
print(df)
```
