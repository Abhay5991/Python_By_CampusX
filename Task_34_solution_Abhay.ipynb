### Q-1 Find out top 10 countries' which have maximum A and D values.

```python
#Q1 Solution using sql workbentch
SELECT country_ab.Country, MAX(country_ab.A) AS Max_A, MAX(country_cd.D) AS Max_D
FROM task_34.country_ab
JOIN task_34.country_cd ON country_ab.Country = country_cd.Country AND country_ab.Edition = country_cd.Edition
GROUP BY country_ab.Country
ORDER BY Max_A DESC, Max_D DESC
LIMIT 10;
```

```python
#Q1 Solution using python
import pymysql
import pandas as pd

# establish a connection to the MySQL database
con = pymysql.connect(host='localhost', user='root', password='campusx', database='task_34')

# execute the SQL query and store the results in a pandas DataFrame
df = pd.read_sql_query("SELECT country_ab.Country, MAX(country_ab.A) AS Max_A, MAX(country_cd.D) AS Max_D \
                        FROM country_ab \
                        JOIN country_cd ON country_ab.Country = country_cd.Country AND country_ab.Edition = country_cd.Edition \
                        GROUP BY country_ab.Country \
                        ORDER BY Max_A DESC, Max_D DESC \
                        LIMIT 10;", con)

# print the resulting DataFrame
print(df)
```

### Q-2 Find out highest CL value for 2020 for every region. Also sort the result in descending order. Also display the CL values in descending order.

```python
#using Sql Workbentch
SELECT c.Region, MAX(cl.CL) AS Highest_CL_2020
FROM task_34.country_ab AS a
JOIN task_34.country_cd AS cd ON a.Country = cd.Country
JOIN task_34.country_cl AS cl ON cd.Country = cl.Country
JOIN task_34.country_efg AS efg ON cl.Country = efg.Country
JOIN task_34.country_ab AS c ON a.Country = c.Country
WHERE cl.Edition = 2020
GROUP BY c.Region
ORDER BY Highest_CL_2020 DESC;
```

```python
#using python
import pymysql

# establish a connection
con = pymysql.connect(host='localhost', user='root', password='campusx', database='task_34')

# execute the query
query = """
    SELECT c.Region, MAX(cl.CL) AS Highest_CL_2020
    FROM task_34.country_ab AS a
    JOIN task_34.country_cd AS cd ON a.Country = cd.Country
    JOIN task_34.country_cl AS cl ON cd.Country = cl.Country
    JOIN task_34.country_efg AS efg ON cl.Country = efg.Country
    JOIN task_34.country_ab AS c ON a.Country = c.Country
    WHERE cl.Edition = 2020
    GROUP BY c.Region
    ORDER BY Highest_CL_2020 DESC;
"""
df = pd.read_sql(query, con)

# display the result
print(df)
```

```python
#after fatching data from sql workbentch
import pandas as pd
import pymysql

# establish connection
con = pymysql.connect(host='localhost', user='root', password='campusx', database='task_34')

# Read the SQL tables into dataframes
ab_df = pd.read_sql_query('SELECT * FROM country_ab', con)
cd_df = pd.read_sql_query('SELECT * FROM country_cd', con)
cl_df = pd.read_sql_query('SELECT * FROM country_cl', con)
efg_df = pd.read_sql_query('SELECT * FROM country_efg', con)

# Merge the dataframes
merged_df = pd.merge(ab_df, cd_df, on=['Country', 'Edition'], how='outer')
merged_df = pd.merge(merged_df, cl_df, on=['Country', 'Edition'], how='outer')
merged_df = pd.merge(merged_df, efg_df, on=['Country', 'Edition'], how='outer')

# Filter the dataframe for 2020
merged_2020 = merged_df[merged_df['Edition'] == 2020]

# Group the dataframe by region and find the maximum CL value
cl_max = merged_2020.groupby('Region')['CL'].max()

# Sort the result in descending order
cl_max = cl_max.sort_values(ascending=False)

# Display the result
print(cl_max)
```

# Q-3 Find top-5 most sold products.

```python
#using Sql workbentch
SELECT p.Name, SUM(s.Quantity) AS TotalQuantity
FROM task_34.Sales s
JOIN task_34.Products p ON s.ProductID = p.ProductID
JOIN task_34.Customers c ON s.CustomerID = c.CustomerID
JOIN task_34.Employees e ON s.SalesPersonID = e.EmployeeID
GROUP BY p.ProductID, p.Name
ORDER BY TotalQuantity DESC
LIMIT 5;
```

```python
import pymysql
import pandas as pd

# Open database connection
con = pymysql.connect(host='localhost', user='root', password='campusx', database='task_34')

# Create cursor object
cur = con.cursor()

# Define SQL query
query = '''
            SELECT p.Name, SUM(s.Quantity) AS TotalQuantity
            FROM Sales s
            JOIN Products p ON s.ProductID = p.ProductID
            JOIN Customers c ON s.CustomerID = c.CustomerID
            JOIN Employees e ON s.SalesPersonID = e.EmployeeID
            GROUP BY p.ProductID, p.Name
            ORDER BY TotalQuantity DESC
            LIMIT 5;
'''


# Execute SQL query and fetch results into a pandas dataframe
df = pd.read_sql(query, con)

# Print the result
print(df)
```

# Q-6: Sales man who has generated most revenue. Show top 5.

```python
#using Sql workbentch
use task_34;
SELECT e.EmployeeID, e.FirstName, e.LastName, ROUND(SUM(s.Quantity * p.Price),2) AS TotalRevenue
FROM Sales s
JOIN Products p ON s.ProductID = p.ProductID
JOIN Employees e ON s.SalesPersonID = e.EmployeeID
JOIN Customers c ON s.CustomerID = c.CustomerID
GROUP BY e.EmployeeID, e.FirstName, e.LastName
ORDER BY TotalRevenue DESC
LIMIT 5;
```

```python
import pandas as pd
import pymysql

# Open database connection
con = pymysql.connect(host='localhost', user='root', password='campusx', database='task_34')

# Define SQL query
query = '''
        SELECT e.EmployeeID, e.FirstName, e.LastName, ROUND(SUM(s.Quantity * p.Price),2) AS TotalRevenue
        FROM Sales s
        JOIN Products p ON s.ProductID = p.ProductID
        JOIN Employees e ON s.SalesPersonID = e.EmployeeID
        JOIN Customers c ON s.CustomerID = c.CustomerID
        GROUP BY e.EmployeeID, e.FirstName, e.LastName
        ORDER BY TotalRevenue DESC
        LIMIT 5;
'''

# Execute SQL query and fetch results into a pandas dataframe
df = pd.read_sql(query, con)

# Print the result
print(df)
```

# Q-7: List all customers who have made more than 10 purchases.

```python
#using sql
SELECT c.CustomerID, c.FirstName, c.LastName, COUNT(s.SalesID) AS TotalPurchases
FROM Customers c
JOIN Sales s ON s.CustomerID = c.CustomerID
GROUP BY c.CustomerID, c.FirstName, c.LastName
HAVING COUNT(s.SalesID) > 10;
```

```python
from sqlalchemy import create_engine
import pandas as pd

# Establish database connection
con = pymysql.connect(host='localhost', user='root', password='campusx', database='task_34')

# Define SQL query
query = '''
            SELECT e.EmployeeID, e.FirstName, e.MiddleInitial, e.LastName, COUNT(DISTINCT c.CustomerID) AS TotalCustomers
            FROM Sales s
            JOIN Employees e ON s.SalesPersonID = e.EmployeeID
            JOIN Customers c ON s.CustomerID = c.CustomerID
            GROUP BY e.EmployeeID, e.FirstName, e.MiddleInitial, e.LastName
            HAVING COUNT(DISTINCT c.CustomerID) > 5
            ORDER BY TotalCustomers DESC;
'''

# Execute SQL query and fetch results into a pandas dataframe
df = pd.read_sql(query, con)

# Print the result
print(df)
```

## Q-9: List all pairs of customers who have made purchases with the same salesperson.

### Answer
### there are no any pairs of customers who have made purchases with the same salesperson

```python
from sqlalchemy import create_engine
import pandas as pd

# Establish database connection
con = pymysql.connect(host='localhost', user='root', password='campusx', database='task_34')

# Define SQL query
query = '''
        SELECT e1.EmployeeID AS SalesPersonID, c1.CustomerID, c1.FirstName AS FirstCustomerFirstName, c1.LastName AS FirstCustomerLastName,
               c2.CustomerID AS OtherCustomerID, c2.FirstName AS OtherCustomerFirstName, c2.LastName AS OtherCustomerLastName
        FROM Sales s1
        JOIN Customers c1 ON s1.CustomerID = c1.CustomerID
        JOIN Employees e1 ON s1.SalesPersonID = e1.EmployeeID
        JOIN Sales s2 ON s1.SalesPersonID = s2.SalesPersonID AND s1.CustomerID < s2.CustomerID
        JOIN Customers c2 ON s2.CustomerID = c2.CustomerID
        GROUP BY e1.EmployeeID, c1.CustomerID, c1.FirstName, c1.LastName, c2.CustomerID, c2.FirstName, c2.LastName
        HAVING COUNT(DISTINCT s1.CustomerID) > 1;
'''

# Execute SQL query and fetch results into a pandas dataframe
df = pd.read_sql(query, con)

# Print the result
print(df)
```
