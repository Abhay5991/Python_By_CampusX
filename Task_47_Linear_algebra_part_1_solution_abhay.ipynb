### Q3. Given the vectors D = [1, 2, 3] and E = [4, 5, 6], calculate the cosine of the angle between them.

````python
import numpy as np

# Define the vectors D and E
D = np.array([1, 2, 3])
E = np.array([4, 5, 6])

# Calculate the dot product of D and E
dot_product = np.dot(D, E)

# Calculate the magnitudes of D and E
magnitude_D = np.linalg.norm(D)
magnitude_E = np.linalg.norm(E)

# Calculate the cosine of the angle between D and E
cosine_theta = dot_product / (magnitude_D * magnitude_E)

print(cosine_theta)
````

### Q6. Given the vectors I = [1, 2, 3] and J = [4, 5, 6], calculate the projection of vector I onto vector J.

````python
import numpy as np

# Define the vectors I and J
I = np.array([1, 2, 3])
J = np.array([4, 5, 6])

# Calculate the squared magnitude of J
J_mag_squared = np.dot(J, J)

# Calculate the dot product of I and J
I_dot_J = np.dot(I, J)

# Calculate the projection of I onto J
proj_I_on_J = (I_dot_J / J_mag_squared) * J

print(proj_I_on_J)
````

````python
import plotly.graph_objs as go
import numpy as np

# Define the vectors I and J
I = np.array([1, 2, 3])
J = np.array([4, 5, 6])

# Calculate the squared magnitude of J
J_mag_squared = np.dot(J, J)

# Calculate the dot product of I and J
I_dot_J = np.dot(I, J)

# Calculate the projection of I onto J
proj_I_on_J = (I_dot_J / J_mag_squared) * J

# Create traces for the vectors and projection
I_trace = go.Scatter3d(x=[0, I[0]], y=[0, I[1]], z=[0, I[2]], name='I', marker=dict(size=8))
J_trace = go.Scatter3d(x=[0, J[0]], y=[0, J[1]], z=[0, J[2]], name='J', marker=dict(size=8))
proj_trace = go.Scatter3d(x=[0, proj_I_on_J[0]], y=[0, proj_I_on_J[1]], z=[0, proj_I_on_J[2]], name='Projection of I on J', marker=dict(size=8))

# Create layout for the plot
layout = go.Layout(scene=dict(xaxis=dict(title='X axis'), yaxis=dict(title='Y axis'), zaxis=dict(title='Z axis')))

# Create figure and add traces to it
fig = go.Figure(data=[I_trace, J_trace, proj_trace], layout=layout)

# Show the plot
fig.show()
````

### Q7. Consider the 2D - position vector a = (1, 3). If b is the point (5, 4), find the vector `ab` and the vector sum `a + ab`. Provide a sketch plot.
Note: Search for quiver, to plot an arrow.

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the position vectors a and b
a = np.array([1, 3])
b = np.array([5, 4])

# Calculate the vector ab
ab = b - a

# Calculate the vector sum a + ab
a_plus_ab = a + ab

# Plot the vectors
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0, 0, ab[0], ab[1], angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0, 0, a_plus_ab[0], a_plus_ab[1], angles='xy', scale_units='xy', scale=1, color='g')
plt.xlim(0, 7)
plt.ylim(0, 7)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['a', 'ab', 'a + ab'])
plt.grid()
plt.show()
````

### Q8. Given two 2D vectors `u` and `v` where `u = [-1, 4]` and `v = [3, -2]`, find the dot product `u . v`, the angle between `u` and `v`, and the projection of `u` onto `v`.

````python
import numpy as np

# Define the vectors u and v
u = np.array([-1, 4])
v = np.array([3, -2])

# Calculate the dot product of u and v
dot_uv = np.dot(u, v)

# Calculate the magnitudes of u and v
mag_u = np.linalg.norm(u)
mag_v = np.linalg.norm(v)

# Calculate the angle between u and v in radians
cos_theta = dot_uv / (mag_u * mag_v)
theta = np.arccos(cos_theta)

# Calculate the projection of u onto v
proj_u_v = (dot_uv / mag_v**2) * v

# Print the results
print("The dot product of u and v is:", dot_uv)
print("The angle between u and v is:", np.degrees(theta), "degrees")
print("The projection of u onto v is:", proj_u_v)
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors u and v
u = np.array([-1, 4])
v = np.array([3, -2])

# Calculate the dot product of u and v
dot_uv = np.dot(u, v)

# Calculate the magnitudes of u and v
mag_u = np.linalg.norm(u)
mag_v = np.linalg.norm(v)

# Calculate the angle between u and v in radians
cos_theta = dot_uv / (mag_u * mag_v)
theta = np.arccos(cos_theta)

# Calculate the projection of u onto v
proj_u_v = (dot_uv / mag_v**2) * v

# Plot the vectors
fig, ax = plt.subplots()
ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(0, 0, proj_u_v[0], proj_u_v[1], angles='xy', scale_units='xy', scale=1, color='g')
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Vectors')
ax.legend(['u', 'v', 'proj_u_v'])

# Print the results
print("The dot product of u and v is:", dot_uv)
print("The angle between u and v is:", np.degrees(theta), "degrees")
print("The projection of u onto v is:", proj_u_v)

plt.show()
````

### Q9. Given a hyperplane defined by the equation 3x - 2y + z = 1, find two different points that lie on this hyperplane.

````python
import numpy as np

# Define the hyperplane
a, b, c, d = 3, -2, 1, 1

# Choose arbitrary values for x and y
x1, y1 = 0, 1
x2, y2 = 1, 0

# Solve for z
z1 = (d - a*x1 - b*y1) / c
z2 = (d - a*x2 - b*y2) / c

# Print the two points
print("The two points on the hyperplane are:")
print("({0}, {1}, {2})".format(x1, y1, z1))
print("({0}, {1}, {2})".format(x2, y2, z2))
````

### Q10. Find a nearest point from a given array of points to a point P1.
P1 - (3,9,3)

Array of points :
``` 
[[1, 8, 5],
[8, 8, 2],
[7, 9, 1],
[5, 6, 3],
[9, 1, 2],
[6, 5, 9],
[5, 8, 6],
[3, 4, 8],
[6, 8, 4],
[7, 2, 7]]
```




````python
import numpy as np

# Define the array of points
points = np.array([[1, 8, 5],
                   [8, 8, 2],
                   [7, 9, 1],
                   [5, 6, 3],
                   [9, 1, 2],
                   [6, 5, 9],
                   [5, 8, 6],
                   [3, 4, 8],
                   [6, 8, 4],
                   [7, 2, 7]])

# Define the point P1
P1 = np.array([3, 9, 3])

# Calculate the distance between P1 and each point in the array
distances = np.sqrt(np.sum((points - P1)**2, axis=1))

# Find the index of the nearest point
index = np.argmin(distances)

# Get the nearest point from the array
nearest_point = points[index]

# Print the result
print("The nearest point to P1 is: ", nearest_point)
````
