### Answer a:- The mean is 50 and the standard deviation is 10.

### Answer b:
To find the value of x that has a z-score of 1.4, we can use the formula:

z = (x - mean) / standard deviation

rearranging this formula to solve for x, we get:

x = z * standard deviation + mean

Substituting z = 1.4, mean = 50, and standard deviation = 10, we get:

x = 1.4 * 10 + 50 = 64

Therefore, the value of x that has a z-score of 1.4 is 64.

### Answer c:
To find the z-score that corresponds to x = 30, we can use the same formula as in part (b), but solve for z instead:

z = (x - mean) / standard deviation

Substituting x = 30, mean = 50, and standard deviation = 10, we get:

z = (30 - 50) / 10 = -2

Therefore, the z-score that corresponds to x = 30 is -2.
z(-2,0) is 0.0228 which is 2.28%

### Answer c:
- The difference between positive and negative z-values is simply the direction from the mean that the corresponding x-value lies. 
- A positive z-value means that the corresponding x-value is above the mean, while a negative z-value means that the corresponding x-value is below the mean. 
- In other words, positive z-values correspond to values in the upper tail of the distribution, while negative z-values correspond to values in the lower tail.

### Q. Which feature is the best predictor of the wine class in the wine dataset? (Using pdf plot)

````python
# code here
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine

data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

sns.set_style('whitegrid')
sns.set_palette('colorblind')

for feature in data.feature_names:
    fig, ax = plt.subplots(figsize=(8,6))
    for target in df['target'].unique():
        sns.distplot(df[feature][df['target']==target], label=f'Class {target}')
    ax.set_title(f'{feature} Distribution')
    ax.set_xlabel(feature)
    ax.set_ylabel('Density')
    ax.legend()
    plt.show()
````
