### Q-7: Given a probability density function f(x) = 2x for 0 <= x <= 1 and f(x) = 0 otherwise, compute the cumulative distribution function F(x) and plot it. Use this to find the probability that X is greater than 0.5. 

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the probability density function
def pdf(x):
    return 2*x if 0 <= x <= 1 else 0

# Define the cumulative distribution function
def cdf(x):
    if x <= 0:
        return 0
    elif x >= 1:
        return 1
    else:
        return x**2

# Plot the PDF and CDF
x = np.linspace(-0.2, 1.2, 500)
y_pdf = [pdf(xi) for xi in x]
y_cdf = [cdf(xi) for xi in x]
plt.plot(x, y_pdf, label='PDF')
plt.plot(x, y_cdf, label='CDF')
plt.legend()

# Find the probability that X is greater than 0.5
prob = 1 - cdf(0.5)
print('Probability that X is greater than 0.5:', prob)
```

### Q-8: In a manufacturing process, the thickness of a certain material is known to be normally distributed with a mean of 1.2 mm and a standard deviation of 0.05 mm. What is the probability density function of the thickness? Plot the PDF and use it to compute the probability that the thickness is between 1.1 mm and 1.3 mm.

```python
# answer here
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the mean and standard deviation of the thickness distribution
mu = 1.2
sigma = 0.05

# Define the range of thickness values to plot the PDF
x = np.linspace(1.0, 1.4, 100)

# Compute the PDF of the thickness distribution
pdf = norm.pdf(x, mu, sigma)

# Plot the PDF
plt.plot(x, pdf)
plt.xlabel('Thickness (mm)')
plt.ylabel('Probability density')
plt.title('PDF of thickness distribution')

# Compute the probability that the thickness is between 1.1 mm and 1.3 mm
prob = norm.cdf(1.3, mu, sigma) - norm.cdf(1.1, mu, sigma)
print('Probability that the thickness is between 1.1 mm and 1.3 mm:', prob)
```

### Q-9 : A data scientist is investigating the distribution of customer ages in a retail store. She collects a sample of 100 ages and estimates the probability density function using kernel density estimation. What bandwidth should she choose to obtain the best estimate?

```python
# answer here
import numpy as np
from scipy.stats import gaussian_kde
import matplotlib.pyplot as plt

# Generate some sample data
ages = np.random.normal(30, 10, size=100)

# Estimate the kernel density function with different bandwidths
bandwidths = np.arange(0.5, 5, 0.1)
kdes = [gaussian_kde(ages, bw_method=bw) for bw in bandwidths]

# Compute the log-likelihood of the data for each bandwidth
log_likelihoods = [kde.logpdf(ages).sum() for kde in kdes]

# Choose the bandwidth with the highest log-likelihood
best_bandwidth = bandwidths[np.argmax(log_likelihoods)]

# Plot the estimated density function with the best bandwidth
kde = gaussian_kde(ages, bw_method=best_bandwidth)
xs = np.linspace(0, 60, 100)
plt.plot(xs, kde(xs), label='KDE estimate (bw={:.2f})'.format(best_bandwidth))
plt.hist(ages, density=True, alpha=0.5, label='Sample data')
plt.legend()
plt.show()
```
