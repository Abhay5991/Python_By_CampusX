### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China. 

```python
polluted_df = pd.read_csv("/content/drive/MyDrive/CampusX/PANDAS/datasets-session-23/most-polluted-cities.csv")
polluted_df.head(2)
```

```python
# read the dataset
polluted_df = pd.read_csv("/content/drive/MyDrive/CampusX/PANDAS/datasets-session-23/most-polluted-cities.csv")

# filter the data for Iran and China
iran_data = polluted_df[polluted_df["Country"] == "Iran"]
china_data = polluted_df[polluted_df["Country"] == "China"]

# calculate the sum of PM2.5 for each year
iran_sum = iran_data.groupby("Year").sum()["PM2.5"]
china_sum = china_data.groupby("Year").sum()["PM2.5"]

# create a line plot
plt.plot(iran_sum.index, iran_sum.values, label="Iran")
plt.plot(china_sum.index, china_sum.values, label="China")
plt.xlabel("Year")
plt.ylabel("PM2.5")
plt.legend()
plt.show()
```

```python
#Using Plotly
import pandas as pd
import plotly.graph_objects as go

# read the dataset
polluted_df = pd.read_csv("/content/drive/MyDrive/CampusX/PANDAS/datasets-session-23/most-polluted-cities.csv")

# filter the data for Iran and China
iran_data = polluted_df[polluted_df["Country"] == "Iran"]
china_data = polluted_df[polluted_df["Country"] == "China"]

# calculate the sum of PM2.5 for each year
iran_sum = iran_data.groupby("Year").sum()["PM2.5"]
china_sum = china_data.groupby("Year").sum()["PM2.5"]

# create a line plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=iran_sum.index, y=iran_sum.values, name="Iran"))
fig.add_trace(go.Scatter(x=china_sum.index, y=china_sum.values, name="China"))
fig.update_layout(xaxis_title="Year", yaxis_title="PM2.5")
fig.show()
```

###`Problem-6`
Show a line plot of Total Profit for each month with below styling.
* Dotted Line
* Line Color Blue
* Show Legend at top left
* Circle Marker

```python
sales_df = pd.read_csv("/content/drive/MyDrive/CampusX/PANDAS/datasets-session-23/company_sales_data.csv")
sales_df.head()
```

```python
# code here

# group the data by month and calculate the total profit for each month
monthly_profit = sales_df.groupby("month_number")["total_profit"].sum()

# create a line plot
plt.plot(monthly_profit.index, monthly_profit.values, '-o', linestyle='dotted', color='blue', marker='o')
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.legend(["Total Profit"], loc="upper left")
plt.show()
```

```python
#using plotly
import pandas as pd
import plotly.graph_objects as go

# group the data by month and calculate the total profit for each month
monthly_profit = sales_df.groupby("month_number")["total_profit"].sum()
# create a line plot
fig = go.Figure()
fig.add_trace(go.Scatter(x=monthly_profit.index, 
                         y=monthly_profit.values,
                         mode='lines+markers',
                         marker=dict(size=10, 
                                     symbol='circle', 
                                     color='blue'), 
                         line=dict(width=2, dash='dot')))
fig.update_layout(xaxis_title="Month",
                  yaxis_title="Total Profit",
                  legend=dict(x=0, y=1.0, 
                  traceorder="normal",
                  font=dict(family="sans-serif", size=12)))
fig.show()
```

###`Problem-9` Show Quarter wise Sales data for all products as multi Bar chart.


```python
sales_df
```

```python
import matplotlib.pyplot as plt

# Group the data by quarter
sales_df['quarter'] = (sales_df['month_number']-1)//3+1
quarters = sales_df.groupby(['quarter'])['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer'].sum()

# Create the bar chart
barWidth = 0.15
r1 = np.arange(len(quarters))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]
r6 = [x + barWidth for x in r5]

plt.bar(r1, quarters['facecream'], color='#7f6d5f', width=barWidth, edgecolor='white', label='Face Cream')
plt.bar(r2, quarters['facewash'], color='#557f2d', width=barWidth, edgecolor='white', label='Face Wash')
plt.bar(r3, quarters['toothpaste'], color='#2d7f5e', width=barWidth, edgecolor='white', label='Toothpaste')
plt.bar(r4, quarters['bathingsoap'], color='#FF9933', width=barWidth, edgecolor='white', label='Bath Soap')
plt.bar(r5, quarters['shampoo'], color='#FF9999', width=barWidth, edgecolor='white', label='Shampoo')
plt.bar(r6, quarters['moisturizer'], color='#FF3333', width=barWidth, edgecolor='white', label='Moisturizer')

plt.xticks([r + barWidth for r in range(len(quarters))], quarters.index)

plt.xlabel("Quarter")
plt.ylabel("Units Sold")
plt.legend()
plt.title('Quarterly Sales Data')
plt.show()
```
